# Download Embedding Model Separately

This notebook downloads the all-MiniLM-L6-v2 model with proper threading controls to avoid mutex deadlock.

Run this once, then the AI server can use the cached model.

```python
# Set environment variables BEFORE importing sentence_transformers
import os
os.environ['OMP_NUM_THREADS'] = '1'
os.environ['TOKENIZERS_PARALLELISM'] = 'false'
os.environ['MKL_NUM_THREADS'] = '1'
os.environ['NUMEXPR_NUM_THREADS'] = '1'

print("Environment configured for single-threaded execution")
print(f"OMP_NUM_THREADS: {os.environ['OMP_NUM_THREADS']}")
print(f"TOKENIZERS_PARALLELISM: {os.environ['TOKENIZERS_PARALLELISM']}")
```

```python
# Download and load the model
from sentence_transformers import SentenceTransformer
import time

print("Starting model download/load...")
start = time.time()

model = SentenceTransformer('all-MiniLM-L6-v2', device='cpu')

elapsed = time.time() - start
print(f"✅ Model loaded successfully in {elapsed:.1f} seconds")
print(f"Model dimension: {model.get_sentence_embedding_dimension()}")
```

```python
# Test the model with a sample encoding
test_text = "Hello, this is a test sentence for embedding."

print("Testing model encoding...")
start = time.time()

embedding = model.encode([test_text], normalize_embeddings=False)[0]

elapsed = time.time() - start
print(f"✅ Encoding successful in {elapsed*1000:.1f}ms")
print(f"Embedding shape: {embedding.shape}")
print(f"First 5 values: {embedding[:5]}")
```

```python
# Check cache location
import pathlib
cache_dir = pathlib.Path.home() / '.cache' / 'huggingface' / 'hub'
model_cache = list(cache_dir.glob('*all-MiniLM-L6-v2*'))

print(f"\nModel cached at:")
for path in model_cache:
    size_mb = sum(f.stat().st_size for f in path.rglob('*') if f.is_file()) / (1024*1024)
    print(f"  {path}")
    print(f"  Size: {size_mb:.1f} MB")
```

## Alternative: Download to Local Directory

If you want to download the model to a specific local directory (not cache), run the cell below:

```python
# Download model to local directory
local_model_path = './models/all-MiniLM-L6-v2'

print(f"Downloading model to: {local_model_path}")
os.makedirs(local_model_path, exist_ok=True)

# This will download to the specified path
local_model = SentenceTransformer('all-MiniLM-L6-v2', device='cpu', cache_folder='./models')
local_model.save(local_model_path)

print(f"✅ Model saved to {local_model_path}")
print(f"\nTo use this in your code, load with:")
print(f"  model = SentenceTransformer('{local_model_path}')")
```
